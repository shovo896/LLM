```python
from langchain_core.prompts import PromptTemplate

template = """I want you act as a phd holder in computer science. I will ask you questions and you 
will answer them as if you were a phd holder in computer science. Do not write explanations. Only answer the question.
My first question is {question}"""

prompt = PromptTemplate(
    input_variables=["question"],
    template=template,
)
prompt.format(question="What is the difference between a list and a tuple in Python?")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0.9)
chain = prompt | llm

response = chain.invoke({"question": "What is the difference between a list and a tuple in Python?"})
response.content
```

```python
## build a code generator that takes a description of a function and generates the code for it.
from langchain import PromptTemplate
code_template = """I want you to act as a code generator. I will describe a function and you will generate the code for it.
My first description is {description}"""
language = "Python"
code_prompt = PromptTemplate(
    input_variables=["description"],
    template=code_template,
)

code_prompt.format(description="a function that takes a list of numbers and returns the sum of the numbers in the list.")
```

```python
chain2 = code_prompt | llm
response2 = chain2.invoke({
    "description": "a function that takes a list of numbers and returns the sum of the numbers in the list."
})
response2.content
```
