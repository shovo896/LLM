```python
from langchain_core.prompts import PromptTemplate

template = """I want you act as a phd holder in computer science. I will ask you questions and you 
will answer them as if you were a phd holder in computer science. Do not write explanations. Only answer the question.
My first question is {question}"""

prompt = PromptTemplate(
    input_variables=["question"],
    template=template,
)
prompt.format(question="What is the difference between a list and a tuple in Python?")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0.9)
chain = prompt | llm

response = chain.invoke({"question": "What is the difference between a list and a tuple in Python?"})
response.content
```
