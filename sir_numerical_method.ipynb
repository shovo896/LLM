```python
import math
x_values = [10, 20, 30, 40, 50]
y_values = [46, 66, 81, 93, 101]
x = 15
x0 = x_values[0]
h = x_values[1] - x_values[0]
p = (x - x0) / h
```

```python
difference_table = [y_values]
for order in range(1, len(y_values)):
    previous_column = difference_table[order - 1]
    current_column = []
    for i in range(len(previous_column) - 1):
        difference = previous_column[i + 1] - previous_column[i]
        current_column.append(difference)
    difference_table.append(current_column)
result = y_values[0]
for order in range(1, len(difference_table)):
    p_term = 1
    for i in range(order):
        p_term *= (p - i)
    term = (p_term / math.factorial(order)) * difference_table[order][0]
    result += term
print(f"f(15) = {result:.4f}")
```
